# Histogram in Matplotlib

### A histogram in Matplotlib is a type of graph used to show the distribution of numerical (continuous) data. It divides the data into intervals (called bins) and shows how many values fall into each interval using bars.

### PLoting Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

### Data Generating

```python
ml_students_age = np.random.randint(18,45,(100))
py_students_age = np.random.randint(15,40,(100))
print(ml_students_age)
print(py_students_age)
```

###  ML Students Age Histogram

```python
plt.hist(ml_students_age,label = "ML_Student" ,color = "b",rwidth =0.9,histtype='bar')
plt.show()
```

```python
plt.hist(ml_students_age)

plt.title("ML Students age histograms")
plt.xlabel("Students age category")
plt.ylabel("No. Students age")

plt.show()
```

### Bins

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_students_age, bins)

plt.title("ML Students age histograms")
plt.xlabel("Students age category")
plt.ylabel("No. Students age")

plt.show()
```

### r width

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_students_age, bins, rwidth=0.8)    

plt.title("ML Students age histograms")
plt.xlabel("Students age category")
plt.ylabel("No. Students age")

plt.show()
```

### histtype : {'bar','barstacked','step','stepfilled'}

```python
bins = [15,20,25,30,35,40,45]
plt.hist(ml_students_age, bins, rwidth=0.8 ,histtype = "step")    

plt.title("ML Students age histograms")
plt.xlabel("Students age category")
plt.ylabel("No. Students age")

plt.show()
```

###   histtype="bar"

```python
import matplotlib.pyplot as plt

ml_students_age = [18, 20, 22, 25, 30, 35, 40, 23, 27, 29, 31]


plt.hist(ml_students_age, bins, rwidth=0.8, histtype="bar")

plt.title("ML Students Age Histogram")
plt.xlabel("Students Age Category")
plt.ylabel("Number of Students")

plt.show()
```

### Horizontal Orientation

```python
ml_students_age = [18, 20, 22, 25, 30, 35, 40, 23, 27, 29, 31]



plt.hist(ml_students_age, bins, rwidth=0.8, histtype="bar", orientation='horizontal')

plt.title("ML Students Age Histogram")
plt.xlabel("Students Age Category")
plt.ylabel("Number of Students")

plt.show()
```

### Color Parameters

```python
ml_students_age = [18, 20, 22, 25, 30, 35, 40, 23, 27, 29, 31]


plt.hist(ml_students_age, bins, rwidth=0.8, histtype="bar", 
         orientation='horizontal', color ="magenta")

plt.title("ML Students Age Histogram")
plt.xlabel("Students Age Category")
plt.ylabel("Number of Students")

plt.show()
```

```python
ml_students_age = [18, 20, 22, 25, 30, 35, 40, 23, 27, 29, 31]


plt.hist(ml_students_age, bins, rwidth=0.8, histtype="bar", 
         orientation='horizontal', color ="red")

plt.title("ML Students Age Histogram")
plt.xlabel("Students Age Category")
plt.ylabel("Number of Students")

plt.show()
```

### Label Parameter

```python
ml_students_age = [18, 20, 22, 25, 30, 35, 40, 23, 27, 29, 31]
plt.hist(ml_students_age, bins, rwidth=0.8, histtype="bar", 
         orientation='horizontal', color ="magenta", label = "ML Student")
plt.title("ML Students Age Histogram")
plt.xlabel("Students Age Category")
plt.ylabel("Number of Students")

plt.legend()

plt.show()
```

### Comparative Histogram of ML and Python Students Age Distribution

```python
import matplotlib.pyplot as plt

ml_students_age = [18, 20, 22, 25, 30, 35, 40, 23, 27, 29, 31]
py_students_age = [19, 21, 24, 26, 28, 32, 36, 38, 41, 22]

bins = [15,20,25,30,35,40,45]

plt.hist(ml_students_age, bins, rwidth=0.8, histtype="bar",
         orientation='horizontal', color="magenta", label="ML Student")

plt.hist(py_students_age, bins, rwidth=0.8, histtype="bar",
         orientation='horizontal', color="grey", label="Py Student")

plt.title("Students Age Histogram")
plt.xlabel("Number of Students")
plt.ylabel("Age Category")

plt.legend()
plt.show()
```

### Vertical Comparison of Histrogram

```python
ml_students_age = np.random.randint(18,45, (100))
py_students_age = np.random.randint(15,40, (100))
bins = [15,20,25,30,35,40,45]

plt.hist([ml_students_age, py_students_age], bins, rwidth=0.8, histtype = "bar",
         orientation='vertical', color = ["magenta", "grey"], label = ["ML Student", "Py Student"])

plt.title("ML & Py Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```

### Plt.figure(figsize)         ( Increase size of bar )

```python
bins = [15,20,25,30,35,40,45]
plt.figure(figsize = (18,10))

plt.hist([ml_students_age, py_students_age], bins, rwidth=0.8, histtype="bar",
         orientation='vertical', color= ["magenta","grey"] , label=["py Student",])

#plt.hist(py_students_age, bins, rwidth=0.8, histtype="bar",
 #        orientation='horizontal', color="grey", label="Py Student")

plt.title("Students Age Histogram")
plt.xlabel("Number of Students")
plt.ylabel("Age Category")

plt.legend()
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")

bins = [15,20,25,30,35,40,45]
plt.figure(figsize = (18,10))

plt.hist([ml_students_age, py_students_age], bins, rwidth=0.8, histtype="bar",
         orientation='vertical', color= ["magenta","grey"] , label=["py Student",])

#plt.hist(py_students_age, bins, rwidth=0.8, histtype="bar",
 #        orientation='horizontal', color="grey", label="Py Student")

plt.title("Students Age Histogram")
plt.xlabel("Number of Students")
plt.ylabel("Age Category")

plt.legend()
plt.show()
```
