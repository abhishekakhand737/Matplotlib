## Showing image using Matplotlib imshow()

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

### Reading Image
##### (use mpimg.imread() method)

```python
img = mpimg.imread("pie_char.png") #address of image location
print(img)
```

#### Image Type

```python
type(img)
```

#### Image Shape

```python
img.shape
```

#### Image Dimension

```python
img.ndim
```

###  Show Image using matplotlib imshow()

```python
plt.imshow(img)
plt.show()
```

### Show image without axis

```python
plt.axis("off")
plt.imshow(img)
plt.show()
```

### Show Image with Colorbar

```python
plt.figure(figsize=(15,8))
plt.axis("off")
plt.imshow(img)
plt.colorbar() # Show color bar of image.
plt.show()
```

### Showing Image with Color map Parameter

```python
single_channel = img[:,:,1]
plt.figure(figsize=(15,8))
plt.axis("off")
plt.imshow(single_channel, cmap = "hot") 
plt.colorbar()
plt.show()
```

```python
single_channel = img[:,:,2]
plt.figure(figsize=(15,8))
plt.axis("off")
plt.imshow(single_channel, cmap = "hot") 
plt.colorbar()
plt.show()
```
