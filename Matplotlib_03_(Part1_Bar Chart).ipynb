# Bar Chart in Matplotlib                                   

 A bar chart is a graphical representation of data using rectangular bars.
The length or height of each bar represents the value or frequency of a category.

### Ploting Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np  
from matplotlib import style
```

### Sample Datasets 

```python
# Datasets of " Ai Edu  class"
#Classes = ["Python","R","AI","ML","DS"]

classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [30, 10, 20, 25, 10] #out of 100 student in each class
class2_students = [40, 5, 20, 20, 10]
class3_students = [35, 5, 30, 15, 15]
```

### Bar Chart of different Classes

```python
plt.bar(classes, class1_students)
plt.show()
```

### Horizontal Plotting Bar

```python
plt.barh(classes,class1_students)
plt.show()
```

## Different Parameter of Bar chart

### Width Parameter (bar size 20% in below example)

```python
plt.bar(classes, class1_students, width = 0.2)   
```

### Align Parameter

```python
plt.bar(classes, class1_students,width =0.2, align = "edge")   
```

### Color Parameter

```python
plt.bar(classes, class1_students, width = 0.2, color = "grey")   
```

### Edge Parameter

```python
plt.bar(classes, class1_students, width = 0.2, color = "grey",
       edgecolor = "red")   
```

### linewidth Method

plt.bar(classes, class1_students, width = 0.2, color = "grey",
       edgecolor = "red", linewidth = 5)   

### Alpa (0-1)

```python
plt.bar(classes, class1_students, width = 0.2, color = "grey",
       edgecolor = "red", linewidth = 5,alpha = 0.6)   
```

### Linestyle

```python
plt.bar(classes, class1_students, width = 0.2, color = "grey",
       edgecolor = "red", linewidth = 5,alpha = 0.6, linestyle ="--")   
```

### All Parameter 

```python
plt.bar(classes, class1_students,
        width=0.2,
        align="edge",
        color="grey",
        edgecolor='red',
        linewidth=5,
        alpha=0.6,
        linestyle="--",
        label="Class 1 Students")

plt.show()
```

### Visible Parameter (It is used to hide bar chart)

```python
plt.bar(classes, class1_students,width=0.2,color="grey",edgecolor="red",linewidth=5,alpha=0.6,linestyle="--",label="Class 1 Students",visible=False)   

plt.legend()
plt.show()
```
