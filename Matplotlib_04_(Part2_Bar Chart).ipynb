# Bar Chart in Matplotlib          (Part_2)                        

```python
import matplotlib.pyplot as plt
import numpy as np  
from matplotlib import style
```

```python
classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [30, 10, 20, 25, 10] #out of 100 student in each class
class2_students = [40, 5, 20, 20, 10]
class3_students = [35, 5, 30, 15, 15]
```

```python
plt.bar(classes, class1_students, width = 0.2, color = "grey",
       edgecolor = "red", linewidth = 5,alpha = 0.6, linestyle ="--", label ="Class 1 Students")   
```

### Increase size of bar chart by width 

```python
plt.bar(classes, class1_students, width = 0.6, color = "grey",
       edgecolor = "red", linewidth = 5,alpha = 0.6, linestyle ="--", label ="Class 1 Students")   
```

### Color Change

```python
plt.bar(classes, class1_students, width = 0.6, color = "black",
       edgecolor = "red", linewidth = 5,alpha = 0.6, linestyle ="--", label ="Class 1 Students")   
```

### Background  Color Change

```python
style.use("ggplot")
plt.bar(classes, class1_students, width = 0.6, color = "black",
       edgecolor = "red", linewidth = 5,alpha = 0.6, linestyle ="--", label ="Class 1 Students")   
```

### Figure Function 
##### (increase size of bar chart)

```python
plt.figure(figsize=(15,9))
plt.bar(classes, class1_students, width = 0.6, color = "black",
       edgecolor = "red", linewidth = 5,alpha = 0.6, linestyle ="--", label ="Class 1 Students")   
```

### Plotting Bar Chart Horizontally

```python
style.use("ggplot")
plt.figure(figsize=(15,9))
plt.barh(classes, class1_students, color = "black", 
       edgecolor = "red", linewidth = 5,alpha = 0.6, linestyle ="--", label ="Class 1 Students")   
```

### Title,X & Y label of Bar Chart

```python
plt.figure(figsize=(15,9))
plt.bar(classes, class1_students,width = 0.6, align ="edge", color = "black", 
       edgecolor = "red", linewidth = 5,alpha = 0.6, linestyle ="--", label ="Class 1 Students")   

plt.title("Bar Chart of Ai Edu class")
plt.xlabel("Classes")
plt.ylabel("No. of Students")
plt.show()
```

### Changing Font Size

```python
plt.figure(figsize=(15,9))
plt.bar(classes, class1_students,width = 0.6, align ="edge", color = "black", 
       edgecolor = "red", linewidth = 5,alpha = 0.6, linestyle ="--", label ="Class 1 Students")   

plt.title("Bar Chart of Ai Edu class",fontsize = 22)
plt.xlabel("Classes",fontsize = 18)
plt.ylabel("No. of Students", fontsize = 18)
plt.show()
```

### Showing Bar Chart of two Classes

```python
plt.figure(figsize=(15,9))
plt.bar(classes, class1_students,width = 0.2, color = "indigo", label ="Class 1 Students") 
plt.bar(classes, class2_students,width = 0.2,color = "green",label ="Class 2 Students") 

plt.title("Bar Chart of Ai Edu class",fontsize = 22)
plt.xlabel("Classes",fontsize = 18)
plt.ylabel("No. of Students", fontsize = 18)
plt.show()
```

### Separate Bar Chart of AI Edu Classes 
##### (Class 1 vs Class 2 Students)

```python
plt.figure(figsize=(15,9))

classes_index = np.arange(len(classes))

width = 0.2

plt.bar(classes_index, class1_students,width , color = "indigo",
        label ="Class 1 Students") 
plt.bar(classes_index + width, class2_students,width ,color = "green",
        label ="Class 2 Students") 

plt.title("Bar Chart of Ai Edu class",fontsize = 22)
plt.xlabel("Classes",fontsize = 18)
plt.ylabel("No. of Students", fontsize = 18)
plt.show()
```

### Xticks Function  ()

```python
#In x-axis previous label changed by new
plt.figure(figsize=(15,9))

classes_index = np.arange(len(classes))

width = 0.2

plt.bar(classes_index, class1_students,width , color = "indigo",
        label ="Class 1 Students") 
plt.bar(classes_index + width, class2_students,width ,color = "green",
        label ="Class 2 Students") 

plt.xticks(classes_index + width, classes,fontsize=13)

plt.title("Bar Chart of Ai Edu class",fontsize = 22)
plt.xlabel("Classes",fontsize = 18)
plt.ylabel("No. of Students", fontsize = 18)
plt.show()
```

### Rotation of label

```python
plt.figure(figsize=(15,9))

classes_index = np.arange(len(classes))

width = 0.2

plt.bar(classes_index, class1_students,width , color = "indigo",
        label ="Class 1 Students") 
plt.bar(classes_index + width, class2_students,width ,color = "green",
        label ="Class 2 Students") 

plt.xticks(classes_index + width, classes,fontsize=13, rotation=25)

plt.title("Bar Chart of Ai Edu class",fontsize = 22)
plt.xlabel("Classes",fontsize = 18)
plt.ylabel("No. of Students", fontsize = 18)
plt.show()
```

### Three Classes Bar

```python
plt.figure(figsize=(15,9))

classes_index = np.arange(len(classes))

width = 0.2

plt.bar(classes_index, class1_students,width , color = "indigo",
        label ="Class 1 Students") 
plt.bar(classes_index + width, class2_students,width ,color = "green",
        label ="Class 2 Students") 
plt.bar(classes_index + width, class3_students,width ,color = "red",
        label ="Class 3 Students") 

plt.xticks(classes_index + width, classes,fontsize=13, rotation=25)

plt.title("Bar Chart of Ai Edu class",fontsize = 22)
plt.xlabel("Classes",fontsize = 18)
plt.ylabel("No. of Students", fontsize = 18)
plt.show()
```

### Plotting of three Class Separate bar chart

```python
plt.figure(figsize=(15,9))

classes_index = np.arange(len(classes))

width = 0.2

plt.bar(classes_index, class1_students,width , color = "indigo",
        label ="Class 1 Students") 
plt.bar(classes_index + width, class2_students,width ,color = "green",
        label ="Class 2 Students") 
plt.bar(classes_index + width + width, class3_students,width ,color = "red",
        label ="Class 3 Students") 

plt.xticks(classes_index + width, classes,fontsize=13, rotation=25)

plt.title("Bar Chart of Ai Edu class",fontsize = 22)
plt.xlabel("Classes",fontsize = 18)
plt.ylabel("No. of Students", fontsize = 18)
plt.show()
```

### Horizontal Plotting of three Class

```python
plt.figure(figsize=(15,9))

classes_index = np.arange(len(classes))

width = 0.2

plt.barh(classes_index, class1_students,width , color = "indigo",
        label ="Class 1 Students") 
plt.barh(classes_index + width, class2_students,width ,color = "green",
        label ="Class 2 Students") 
plt.barh(classes_index + width + width, class3_students,width ,color = "red",
        label ="Class 3 Students") 

plt.yticks(classes_index + width, classes,fontsize=13, rotation=25)

plt.title("Bar Chart of Ai Edu class",fontsize = 22)
plt.ylabel("Classes",fontsize = 18)
plt.xlabel("No. of Students", fontsize = 18)
plt.show()
```

### Legend Function () 
##### (To show label)

```python
plt.figure(figsize=(15,9))

classes_index = np.arange(len(classes))

width = 0.2

plt.barh(classes_index, class1_students,width , color = "indigo",
        label ="Class 1 Students") 
plt.barh(classes_index + width, class2_students,width ,color = "green",
        label ="Class 2 Students") 
plt.barh(classes_index + width + width, class3_students,width ,color = "red",
        label ="Class 3 Students") 

plt.yticks(classes_index + width, classes,fontsize=13, rotation=25)

plt.title("Bar Chart of Ai Edu class",fontsize = 22)
plt.ylabel("Classes",fontsize = 18)
plt.xlabel("No. of Students", fontsize = 18)
plt.legend()
plt.show()
```
