## Matplotlib ( Scatter Plot )

A scatter plot in Matplotlib is a graph that shows the relationship between two variables by plotting individual data points on a coordinate plane.

```python
import matplotlib.pyplot as plt
import pandas as pd  
```

### Scatter Plot: Rating vs Reviews

This scatter plot shows the relationship between the **Rating** and **Reviews** of applications from the Google Play Store dataset.
- **X-axis:** Rating of the app  
- **Y-axis:** Number of reviews received by the app  
- Each point in the graph represents an individual application.

```python
x = df_google_play_store_apps ['Rating']
y = df_google_play_store_apps ['Reviews']
```

```python
plt.scatter(x,y)
```

### Labelling Scatter Plot (X,Y & title)

```python
plt.scatter(x,y)
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

### Red star markers plot

```python

plt.scatter(x,y, color='r',marker = "*", s = 100)
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

### Increase Size of Scatter Plot

```python
plt.figure(figsize = (17,10))
plt.scatter(x,y, color='r')
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (17,10))
plt.scatter(x,y, color='r',marker = "*", s = 100)
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (17,10))
plt.scatter(x,y, color='r',marker = "*", s = 100,alpha=0.5)
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

### Increase Size of Star

```python
plt.figure(figsize = (17,10))
plt.scatter(x,y, color='r',marker = "*", s = 100,alpha=0.5, linewidths=10)
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

```python
plt.figure(figsize = (17,10))
plt.scatter(x,y, color='r',marker = "*", s = 100,alpha=0.5, linewidths=10,
            edgecolors='green')
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
```

### Two Scattered Plot in graph

```python
df_google_play_store_apps = pd.read_csv("C:\\Users\\abhishek\\Downloads\\archive (1)\\googleplaystore.csv", nrows=1000)
x = df_google_play_store_apps ['Rating']
y = df_google_play_store_apps ['Reviews']
plt.figure(figsize = (17,10))
plt.scatter(x,y, color='r',marker = "*", s = 100,alpha=0.5, linewidths=10,
            edgecolors='green')
plt.scatter(x,df_google_play_store_apps["Installs"], color="y" , marker = "o", s = 100,alpha=0.5, linewidths=10,
            edgecolors='c')
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews & Installs")
```

```python
plt.figure(figsize = (17,10))
plt.scatter(x,y, color='r',marker = "*", s = 100,alpha=0.5, linewidths=10,
            edgecolors='red')
plt.scatter(x,df_google_play_store_apps["Installs"], color="y" , marker = "o", s = 100,alpha=0.5, linewidths=10,
            edgecolors='y')
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Installs")
```
