# Matplotlib (Pie Chart)|

A Pie Chart in Matplotlib is used to display the percentage distribution of different categories in a circular form. 📊

```python
import matplotlib.pyplot as plt
```

### Single slice pie chart

```python
classes = [  "Python","R", "Machine Learning","Artificial Intelligence",
    "Data Science"]
class1_students = [45, 15, 35, 25, 30]

plt.pie([1])
```

### Simple Pie Chart Plotting.

```python
classes = [  "Python","R", "Machine Learning","Artificial Intelligence",
    "Data Science"]
class1_students = [45, 15, 35, 25, 30]

plt.pie(class1_students,labels = classes)
plt.show()
```

### Explode Parameter (slice)

```python
classes = ["Python", "R", "Machine Learning", "Artificial Intelligence", "Data Science"]
class1_students = [45, 15, 35, 25, 30]
explode = [0.03, 0, 0.1, 0, 0]
plt.pie(class1_students,labels=classes,explode=explode)
plt.show()
```

### Color Parameter 
##### (coloring pie chart manually)

```python
classes = ["Python", "R", "Machine Learning", "Artificial Intelligence", "Data Science"]
class1_students = [45, 15, 35, 25, 30]
explode = [0.03, 0, 0.1, 0, 0]
colors = ['c', 'b', 'r', 'y', 'g']
plt.pie(class1_students,labels=classes,explode=explode, colors = colors)
plt.show()
```

### Auto Percentage  
##### (Show distribution in Percentage%)

```python
classes = ["Python", "R", "Machine Learning", "Artificial Intelligence", "Data Science"]
class1_students = [45, 15, 35, 25, 30]
explode = [0.03, 0, 0.1, 0, 0]
colors = ['c', 'b', 'r', 'y', 'g']
plt.pie(class1_students, labels=classes, explode=explode, colors = colors, autopct = "%0.2f%%")
plt.show()
```

### Shadow Parameter

```python
classes = ["Python", "R", "Machine Learning", "Artificial Intelligence", "Data Science"]
class1_students = [45, 15, 35, 25, 30]
explode = [0.03, 0, 0.1, 0, 0]
colors = ['c', 'b', 'r', 'y', 'g']
plt.pie(class1_students, labels=classes, explode=explode, colors = colors,
        autopct = "%0.2f%%", shadow = True)
plt.show()
```

### Radius Parameter
##### (Increase size of Pie chart)

```python
classes = ["Python", "R", "Machine Learning", "Artificial Intelligence", "Data Science"]
class1_students = [45, 15, 35, 25, 30]
explode = [0.03, 0, 0.1, 0, 0]
colors = ['c', 'b', 'r', 'y', 'g']
plt.pie(class1_students, labels=classes, explode=explode, colors = colors,
        autopct = "%0.2f%%", shadow = True, radius = 1.4)
plt.show()
```

### Start Angle  (1st pie angle % )

```python
classes = ["Python", "R", "Machine Learning", "Artificial Intelligence", "Data Science"]
class1_students = [45, 15, 35, 25, 30]
explode = [0.03, 0, 0.1, 0, 0]
colors = ['c', 'b', 'r', 'y', 'g']
plt.pie(class1_students, labels=classes, explode=explode, colors = colors,
        autopct = "%0.2f%%", shadow = True, radius = 1.4,startangle = 90)
plt.show()
```

### Texts property (Font Size)

```python
classes = ["Python", "R", "Machine Learning", "Artificial Intelligence", "Data Science"]
class1_students = [45, 15, 35, 25, 30]
explode = [0.03, 0, 0.1, 0, 0]
colors = ['c', 'b', 'r', 'y', 'g']
plt.pie(class1_students, labels=classes, explode=explode, colors = colors,
        autopct = "%0.2f%%", shadow = True, radius = 1.4,startangle = 90)
plt.show()
```

### Colorful Pie Chart.

```python
import matplotlib.pyplot as plt
classes = ["Python", "R", "Machine Learning", "Artificial Intelligence", "Data Science"]

class1_students = [45, 15, 35, 25, 30]

explode = [0.03, 0, 0.1, 0, 0]
# Colors
colors = ['c', 'b', 'r', 'y', 'g']
textprops= {"fontsize":15}
plt.pie(class1_students,
        labels=classes,
        explode=explode,
        colors=colors,
        autopct='%0.2f%%',
        shadow=True,
        radius=1.4,
        startangle=270,
        textprops= textprops)

plt.show()
```

## Styled pie chart with custom wedges
* Percentage Distance   (maintain % distance)
* label distance      (maintain label distance)
* rotate label (rotate label)

```python
wedgeprops = {"linewidth":4, 'width':1, "edgecolor":"k"}
plt.pie(
    class1_students,
    labels=classes,
    explode=explode,
    colors=colors,
    autopct="%0.2f%%",
    pctdistance=0.6,
    shadow=True,
    labeldistance=1.4,
    startangle=270,
    radius=1.1,
    counterclock=True,
    wedgeprops=wedgeprops,
    textprops=textprops,
    center=(2, 3),
    frame=True,
    rotatelabels=True
)
```

```python
wedgeprops = {"linewidth":4, 'width':2, "edgecolor":"k"}
plt.pie(
    class1_students,
    labels=classes,
    explode=explode,
    colors=colors,
    autopct="%0.2f%%",
    pctdistance=0.6,
    shadow=True,
    labeldistance=1.6,
    startangle=270,
    radius=1.1,
    counterclock=True,
    wedgeprops=wedgeprops,
    textprops=textprops,
    center=(2, 3),
    frame=True,
    rotatelabels=True
)
plt.show()
```

```python
wedgeprops = {"linewidth":4, 'width':3, "edgecolor":"k"}
plt.pie(
    class1_students,
    labels=classes,
    explode=explode,
    colors=colors,
    autopct="%0.2f%%",
    pctdistance=0.6,
    shadow=True,
    labeldistance=1.6,
    startangle=270,
    radius=1.1,
    counterclock=True,
    wedgeprops=wedgeprops,
    textprops=textprops,
    center=(2, 3),
    frame=True,
    rotatelabels=True
)
plt.show()
```

### Figsize (small size)

```python
plt.figure(figsize = (6,2))
wedgeprops = {"linewidth":4, 'width':3, "edgecolor":"k"}
plt.pie(
    class1_students,
    labels=classes,
    explode=explode,
    colors=colors,
    autopct="%0.2f%%",
    pctdistance=0.6,
    shadow=True,
    labeldistance=1.4,
    startangle=270,
    radius=0.5,
    counterclock=True,
    wedgeprops=wedgeprops,
    textprops=textprops,
    center=(2, 3),
    frame=True,
    rotatelabels=True
)
plt.show()
```

### Legend

```python
classes = ["Python", "R", "Machine Learning", "Artificial Intelligence", "Data Science"]
class1_students = [45, 15, 35, 25, 30]
explode = [0.03, 0, 0.1, 0, 0]
colors = ['c', 'b', 'r', 'y', 'g']
plt.pie(class1_students, labels=classes, explode=explode, colors = colors,
        autopct = "%0.2f%%", shadow = True, radius = 1.4,startangle = 90)
plt.legend()
plt.show()
```

### Multiple Colorful Circle Pie Chart

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
plt.figure(figsize=(16,9))

colors = ['r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w','r','w']
labels = np.ones(20)

plt.pie([1], colors="k", radius=2.05)
plt.pie(labels, colors=colors, radius=2.0)

plt.pie([1], colors="g", radius=1.8)
plt.pie([1], colors="y", radius=1.6)
plt.pie([1], colors="c", radius=1.3)
plt.pie([1], colors="b", radius=1.1)
plt.pie([1], colors="m", radius=0.9)

plt.pie([1], colors="b", radius=0.31)
plt.pie(labels, colors=colors, radius=0.3)

plt.pie([1], colors="w", radius=0.2)
plt.pie([1], colors="k", radius=0.1)

plt.show()
```
